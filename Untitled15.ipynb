```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
n=15
p=0.2
k=np.arange(0,16)
binomial=stats.binom.pmf(k,n,p)
plt.plot(k, binomial, 'o-',marker = '*')
plt.title('Binomial: n = %i p = %f' %(n,p))
plt.xlabel('Number of attempts')
plt.ylabel('Probability of picking A-Spade')
plt.legend(["A-Spade"])
plt.show()
```

```python
from scipy import stats
import matplotlib.pyplot as plt
rate=4
n=np.arange(0,11)
y=stats.poisson.pmf(n, rate)
plt.plot(n, y,'o-')
plt.title('Poisson: $\lambda$ = %i' %rate)
plt.show()
```
