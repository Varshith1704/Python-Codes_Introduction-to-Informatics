# QUESTION 1

```python
import pandas as pd  
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import numpy as np
data = { 
    #'Food':['Salmon', 'Cottage Cheese', 'Pork', 'Oats','Baked Beans','Chicken','Beef','Peas'],
        'Protein(g)':[31, 18.58, 41.1,25.34,6.2,34.64,40.94,8.13],
        'Fat(g)':[8.2,1.53,25.56,10.35,0.3,1.86,22.52,0.6],
        'Carbohydrates(g)':[0,4.08,0,99.4,17.9,0,0,21.69],
        'Cholesterol(mg)' :[89,26,120,0,0,132,135,0],      
        'Calories':[206,108,406,584,584,165,378,122],
'Food Type':['Non-veg','Veg','Non-veg','Veg','Veg','Non-veg','Non-veg','Veg']
       }  
df = pd.DataFrame(data)  
X = df.iloc[:,:-1].values
y = df.iloc[:, 5].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
classifier = KNeighborsClassifier(n_neighbors=2)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
y_predi = classifier.predict([[33.89,1.77,0.42,95,164]])
print('Fish Food Type : ',y_predi)
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred))
print('Classification_report:')
print(classification_report(y_test, y_pred))
error = []
for i in range(1, 6):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)
    error.append(np.mean(pred_i != y_test))
plt.figure(figsize=(12, 6))
plt.plot(range(1, 6), error, color='red', linestyle='dashed', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Error Rate K Value')
plt.xlabel('K Value')
plt.ylabel('Mean Error')
```

# QUESTION 2

```python
import pandas
from sklearn import linear_model
df = pandas.read_csv("auto-mpg.csv")
X = df[['mpg', 'weight','acceleration']]
y = df['displacement']
regr = linear_model.LinearRegression()
regr.fit(X, y)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
print(regr.fit(X_train,y_train))
y_predicted = regr.predict([[18.5, 3645,16.2]])
print('Displacement for given values is : ',y_predicted)
predicted = regr.predict(X_test)
expected = y_test
print("RMSE: %s" % np.sqrt(np.mean((predicted - expected) ** 2))) 
plt.scatter(expected, predicted) 
plt.show()
```
