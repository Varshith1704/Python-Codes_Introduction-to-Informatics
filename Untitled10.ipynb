```python
import math
class Rectangle():
    def Area(self,l,w):
        v = round(l * w,2)
        return v
    def Perimeter(self,l,w):
        r = round((l * 2) + (w * 2),2)
        print("Perimeter is: ",r)
    def Diagonal(self,l,w):
        s= round(math.sqrt((w**2) + (l**2)),2)
        print("Diagonal is: ",s)
l = int(input("Enter Length: "))
w = int(input("Enter Width: "))
R=Rectangle()
R.Area(l,w)
R.Perimeter(l,w)
R.Diagonal(l,w)
```

```python
class People:# Base class
    def _init_(self, name, surname, number, pincode):# constructor
        self.name = name # instance attribute
        self.surname = surname # instance attribute
        self.number = number # instance attribute
        self.pincode = pincode # instance attribute


class Student(People):# Derived class
    def _init_(self, student_type, *args, **kwargs):# constructor with formal and keyword arguments
        self.student_type = student_type # instance attribute
        self.classes = [] #instance
        super(Student, self)._init_(*args, **kwargs)# constructor of the parent class

    def enroll(self, course):# function
        self.classes.append(course)


class Staffs(People)://class
    def _init_(self, employment_type, *args, **kwargs):# constructor with formal and keyword arguments
        self.employment_type = employment_type
        super(StaffMember, self)._init_(*args, **kwargs)# constructor of the parent class

    
    def salary(self, basicpay):# function
        return(self.basicpay*1000)



class Professor(Staffs):# Derived class
    def _init_(self, *args, **kwargs):# constructor with formal and keyword arguments
        self.courses_taught = [] #instance
        super(Lecturer, self)._init_(*args, **kwargs)

    def assign_teaching(self, course):# Function
        self.courses_taught.append(course)
```
