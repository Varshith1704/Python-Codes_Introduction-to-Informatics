```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import NullFormatter
x1 = [2,10]
y1 = [5,50]
plt.subplot(2, 2, 1)
plt.grid(True)
plt.plot(x1, y1, linewidth=1, color='red')
x2 = np.array([1,5,10,15,20,25])
y2= np.array([10,40,10,60,30,20])
plt.subplot(2, 2, 2)
plt.grid(True)
plt.scatter(x2, y2, marker = 'x')
y3= [10,40,10,60,30]
x3 = [0.5,2, 3, 4,5]
plt.subplot(2, 2, 3)
plt.grid(True)
plt.barh(x3,y3, color='gray')
y4 = [22,32.5,50,95,30]
x4 = [2,4, 6, 8,10]
plt.subplot(2, 2, 4)
plt.grid(True)
plt.bar(x4,y4, color='lightblue')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
data=[22, 10, 40, 18, 10]
label = ["Laptop", "Tablet","Phone","TV", "Computer"]
colors = ( "black", "red", "blue", "darkgreen","purple") 
plt.pie(data, labels=label, colors=colors ,autopct='%1.f%%',startangle=100)
plt.title('Student Usage')
plt.axis('equal')
plt.show()
```
